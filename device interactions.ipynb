```python
for row in data.splitlines():
    print(row.split()[1])
```

```python
for row in data.splitlines():
    row = row.decode("utf-8") # converts byte string to normal string
    if "sublime" in row:
        print(row.split()[1])
        break
```

```python
for row in data.splitlines():
    row = row.decode("utf-8") # converts byte string to normal string
    if "sublime" in row:
        pid = row.split()[1]
        cmd="kill -9 %s"%pid
        os.system(cmd)
        break
```
