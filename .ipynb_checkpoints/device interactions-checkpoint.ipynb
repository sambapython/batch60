```python
for row in data.splitlines():
    row = row.decode("utf-8") # converts byte string to normal string
    if "sublime" in row:
        print(row.split()[1])
        break
```

```python
for row in data.splitlines():
    row = row.decode("utf-8") # converts byte string to normal string
    if "sublime" in row:
        pid = row.split()[1]
        cmd="kill -9 %s"%pid
        os.system(cmd)
        break
```

```python
data1= []
table = bsd.find('table')
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data1.append([ele for ele in cols if ele]) # Get rid of empty values
```

```python
import paramiko
ssh = paramiko.SSHClient()
ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
ssh.connect("localhost",username="khyaathipython",password="python")
```

```python
ssh.exec_command("mkdir local")
```

```python
inp, outp, errp = ssh.exec_command("mkdir local")
print(errp.read())
print(outp.read())
```

```python
inp, outp, errp = ssh.exec_command("mkdir remote")
print(errp.read())
print(outp.read())
```

```python
inp, outp, errp = ssh.exec_command("ls -l")
print(errp.read())
print(outp.read())
```

```python
inp, outp, errp = ssh.exec_command("echo python| sudo -S ls /proc") # python is a password of remote machine..
print(errp.read())
print(outp.read())
```
